```python
import pandas as pd
dframe_wine = pd.read_csv('winequality-red.csv')
```

```python
# Sort the DataFrame by 'alcohol' in descending order
#inplace=True: Modifies the original DataFrame directly. No new DataFrame is created, and no assignment is necessary.
#inplace=False: Returns a new DataFrame with the sorted values, leaving the original DataFrame unchanged.
dframe_wine.sort_values('alcohol', ascending=False, inplace=True)
dframe_wine.head()
```

```python
import numpy as np

def ranker(df):
    # Create a new column 'alc_content_rank' that ranks the wines
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
# Apply the ranker function to each group
dframe_wine = dframe_wine.groupby('quality').apply(ranker)
```

```python
# Display the first few rows of the DataFrame
dframe_wine.head()
```

```python
# Count the number of wines for each quality level
num_of_qual = dframe_wine['quality'].value_counts()
num_of_qual
```

```python
# Filter the DataFrame to show top-ranked wines for each quality level
top_ranked_wines = dframe_wine[dframe_wine.alc_content_rank == 1]

# Display the top-ranked wines
top_ranked_wines
```

```python
# Generate a random dataset with 25 samples
dataset = randn(25)

# Plot a rug plot
sns.rugplot(dataset)
plt.ylim(0, 1)
plt.title('Rug Plot')
plt.show()

# Plot a histogram with rug plot
plt.hist(dataset, alpha=0.3)
sns.rugplot(dataset)
plt.title('Histogram with Rug Plot')
plt.show()
```

```python
# Plotting using lmplot with the wine dataset
# The term lmplot stands for "linear model plot." The name reflects its primary purpose: to visualize the relationship between two variables with the inclusion of a linear regression model (or line) that is fitted to the data.
sns.lmplot(x='fixed acidity', y='volatile acidity', data=dframe_wine,
           scatter_kws={'marker': 'o', 'color': 'indianred'},
           line_kws={'linewidth': 1, 'color': 'blue'})
plt.title('Linear Regression Plot of Fixed Acidity vs Volatile Acidity')
plt.show()
```

```python
# Estimate for tendency for each bin
sns.lmplot(x='chlorides', y='sulphates', data=dframe_wine, x_estimator=np.mean)
"""
x_estimator=np.mean:
The x_estimator parameter specifies a function to apply to the x values within each bin.
np.mean indicates that the mean value of chlorides will be computed for each bin.
This mean value is then used to create the estimated central tendency line in the plot.
"""
plt.title('Linear Regression Plot of Chlorides vs Sulphates with Mean Estimator')
plt.show()
```

```python
sns.lmplot(x='density', y='pH', data=dframe_wine, hue='quality')
plt.title('Linear Regression Plot of Density vs pH with Quality Hue')
plt.show()
```

```python
# Create a figure with two subplots sharing the y-axis
# sharey=True: This ensures that the y-axis is shared between the two subplots, allowing for direct comparison.
fig, (axis1, axis2) = plt.subplots(1, 2, sharey=True, figsize=(12, 6))

# Plot the regression plot on the first subplot (axis1)
# ax=axis1: Plots the regression plot on the first subplot (axis1).
# scatter_kws={'marker': 'o', 'color': 'indianred'}: Customizes the appearance of the scatter points (using circles as markers and the color indianred).
scatter_kws={'marker': 'o', 'color': 'indianred'}: Customizes the appearance of the scatter points (using circles as markers and the color indianred).
sns.regplot(x='fixed acidity', y='volatile acidity', data=dframe_wine, ax=axis1, scatter_kws={'marker': 'o', 'color': 'indianred'})
axis1.set_title('Regression Plot of Fixed Acidity vs Volatile Acidity')

# Plot the violin plot on the second subplot (axis2)
sns.violinplot(x='quality', y='alcohol', data=dframe_wine, palette='Reds_r', ax=axis2)
axis2.set_title('Violin Plot of Quality vs Alcohol Content')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Convert a subset of columns to integers for heatmap (assuming quality and alcohol)
# Basically conversion of decimals
heatmap_data = dframe_wine[['quality', 'alcohol']].astype(int)
```

```python
# Reset index to avoid ambiguity
heatmap_data = heatmap_data.reset_index(drop=True)

"""
# Sample DataFrame
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales': [1500, 1800, 1200, 2100],
    'Expenses': [1000, 1100, 900, 1200]
}

sales_data = pd.DataFrame(data)
print(sales_data)
"""
"""
  Month  Sales  Expenses
0   Jan   1500      1000
1   Feb   1800      1100
2   Mar   1200       900
3   Apr   2100      1200
"""
```

```python
# Pivot the dataframe (simulating a heatmap)
heatmap_dframe = heatmap_data.pivot_table(index='quality', columns='alcohol', aggfunc=len, fill_value=0)
# aggfunc=len: Specifies the aggregation function to be used when aggregating values. In this case, len (which stands for length or count) is used to count occurrences of combinations of 'quality' and 'alcohol' values.
# fill_value=0: Specifies the value to replace missing values (NaN) in the pivot table. Here, it's set to 0, meaning if there are no occurrences for a particular combination of 'quality' and 'alcohol', it will be filled with 0.
```

```python
# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_dframe, annot=True, fmt='d', cmap='coolwarm', cbar=True)
"""
heatmap_dframe: This is the pivot table (previously created) containing counts of combinations of 'quality' and 'alcohol' values.
annot=True: This parameter adds numeric annotations (the counts) to each cell of the heatmap.
fmt='d': This specifies the format of the annotations as integers ('d' stands for integer).
"""
plt.title('Heatmap of Wine Quality vs Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Wine Quality')
plt.show()
```

```python
# Centering heatmap on a specific cell (simulating centering on a particular month and year)
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_dframe, center=heatmap_dframe.loc[5, 10], annot=True, fmt='d', cmap='YlGnBu', cbar=True)
plt.title('Heatmap Centered on Wine Quality 5 and Alcohol Content 10')
plt.xlabel('Alcohol Content')
plt.ylabel('Wine Quality')
plt.show()
```

# GROUPBY ON DATAFRAMES

# Groupby on DataFrames:

Groupby operation in pandas is used to split the data into groups based on some criteria, apply a function to each group independently, and then combine the results into a DataFrame or Series.

# Usage:
It's used for data aggregation, transformation, and analysis tasks, such as calculating group-wise statistics, summarizing data, and performing group-wise operations.


It groups the DataFrame dframe by different columns ('k1', 'k2', data types), calculates group-wise statistics like mean, size, and creates dictionaries containing groups. Additionally, it showcases iterating over groups and accessing specific groups.

```python
#Import Libraries
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
```

```python
#Create DataFrame
dframe = DataFrame({'k1':['X','X','Y','Y','Z'],
                    'k2':['alpha','beta','alpha','beta','alpha'],
                   'dataset1':np.random.randn(5),
                   'dataset2':np.random.randn(5)})
```

```python
#Display DataFrame
dframe
```

```python
#Grouping Data
group1 = dframe['dataset1'].groupby(dframe['k1'])
group1
```

```python
#Calculate Mean
group1.mean()
```

```python
#Grouping with Multiple Columns:
cities = np.array(['NY','LA','LA','NY','NY'])
month = np.array(['JAN','FEB','JAN','FEB','JAN'])
```

```python
dframe['dataset1'].groupby([cities,month]).mean()
```
